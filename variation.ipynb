```python
import requests
from bs4 import BeautifulSoup
import time
import csv

# Base URL of the Google Scholar page
base_url = "https://scholar.google.co.id/citations?view_op=view_org&hl=en&org=3079205246798671920"

# Headers to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

# Function to scrape a single page
def scrape_page(url, page_number):
    print(f"Processing page {page_number}: {url}")
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Extract the desired information
        # Example: Extracting author names
        items = soup.find_all('div', class_='gsc_1usr')
        print(len(items))
        for item in items:
            name = item.find('h3', class_='gs_ai_name').text
            variation_name = item.find('div', class_='gs_ai_aff').text
            #print(name, variation_name)
            with open('v.csv', mode='a', newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                # Write the name and variation_name to the CSV file
                writer.writerow([name, variation_name])
        
        # Find the next page link
        next_button = soup.find('button', {'aria-label': 'Next'})
        if next_button and 'disabled' not in next_button.get('class', []):
            next_url = next_button['onclick'].split("window.location='")[1].split("'")[0]
            next_url = "https://scholar.google.co.id" + next_url.replace("\\x3d", "=").replace("\\x26", "&")
            return next_url
        else:
            return None
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None

# Start scraping from the first page
current_url = base_url
page_number = 1
while current_url:
    current_url = scrape_page(current_url, page_number)
    page_number += 1
    print(f"Page {page_number}: {current_url}")
    time.sleep(2)  # Add a delay between requests to avoid rate limiting
```
